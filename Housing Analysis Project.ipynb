```python
#Read data

df_byu_b = pd.read_csv('https://github.com/bdi475/datasets/raw/main/campustowns-leasing-company-reviews/byu-businesses.csv')
df_byu_r = pd.read_csv('https://github.com/bdi475/datasets/raw/main/campustowns-leasing-company-reviews/byu-reviews.csv')
df_psu_b = pd.read_csv('https://github.com/bdi475/datasets/raw/main/campustowns-leasing-company-reviews/psu-businesses.csv')
df_psu_r = pd.read_csv('https://github.com/bdi475/datasets/raw/main/campustowns-leasing-company-reviews/psu-reviews.csv')
df_uiuc_b = pd.read_csv('https://github.com/bdi475/datasets/raw/main/campustowns-leasing-company-reviews/uiuc-businesses.csv')
df_uiuc_r = pd.read_csv('https://github.com/bdi475/datasets/raw/main/campustowns-leasing-company-reviews/uiuc-reviews.csv')
```

This code is loading six datasets from a GitHub repository via direct links: it retrieves data for businesses and reviews separately from Brigham Young University (BYU), Pennsylvania State University (PSU), and the University of Illinois at Urbana-Champaign (UIUC), resulting in a pair of DataFrames for each university campus—one containing information about leasing companies (businesses) and the other containing reviews of those companies.

```python
#Textual Analysis

pd.set_option('display.max_columns', 100)
import spacy
nlp = spacy.load('en_core_web_sm')
text = 'I love this apartment'
doc = nlp(text)

for token in doc:
    print('------------------')
    print(f'text: {token.text}')
    print(f'lemma: {token.lemma_}')
    print(f'pos: {token.pos_}') # pos_ stands for part-of-speech
    print(f'explain: {spacy.explain(token.pos_)}')
    print(f'is_stop: {token.is_stop}')
```

This code snippet is performing a basic textual analysis on a sample sentence "I love this apartment" using the spaCy Natural Language Processing (NLP) library

```python
cols = ["text", "lemma", "pos", "explain", "is_stop"]
rows = []

for t in doc:
    row = [t.text, t.lemma_, t.pos_, spacy.explain(t.pos_), t.is_stop]
    rows.append(row)

df_tokens = pd.DataFrame(rows, columns=cols)
df_tokens
```

```python
! pip install transformers
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
```

```python
#Test classifier
classifier("These thieves tried to steal my security deposit.")
```

```python
#Create sentiment and score columns within reviews dataframe (filled with NAN values)

df_r['sentiment'] = np.nan
df_r['score'] = np.nan
```

```python
#Populate sentiment and score columns for review dataframe

num_rows = df_r.shape[0]

for i in range(num_rows):
  # grab review text to a variable
  review_text = df_r['review_text'].iloc[i]

  # calculate sentiments for non-missing review texts
  # the model supports up to 512 tokens
  # truncate longer texts
  if pd.notna(review_text):
    result = classifier(
        review_text,
        truncation=True,
        padding=True,
        max_length=512
    )

    df_r.iloc[i, df_r.columns.get_loc('sentiment')] = result[0]['label']
    df_r.iloc[i, df_r.columns.get_loc('score')] = result[0]['score']

  # display progress
  progress_percentage = round((i + 1) / num_rows * 100, 2)
  print(f'{i + 1}/{num_rows} ({progress_percentage}%)', end=' ')

  if (i + 1) % 10 == 0:
    print('')

print('====================')

#Clean output
from IPython.display import clear_output
clear_output()

print('Complete')
```

```python
#Examine different tags within Review Tags Column

column_data = df_b['reviews_tags']
print(column_data)
print(column_data.head())
```

```python
#Visualizatiion 1: Tag Counts per Business Category

included_tags = ['pest control', 'pests', 'prices', 'mold', 'fees', 'management', 'rent']

# Process tags
all_tags = []
for index, row in df_b.iterrows():
    # Check if reviews_tags is a string (and not NaN or other types)
    if isinstance(row['reviews_tags'], str):
        tags = row['reviews_tags'].split(', ')
        for tag in tags:
            # Only include the tag if it's in the list of included tags
            if tag in included_tags:
                all_tags.append({'type': row['type'], 'tag': tag})

# Create a new DataFrame for tags
tags_df = pd.DataFrame(all_tags)

# Count tags per category
tag_counts = tags_df.groupby(['type', 'tag']).size().reset_index(name='count')


# Plot using Plotly
fig = px.bar(tag_counts, x='type', y='count', color='tag', title='Tag Counts per Business Category')

fig.update_layout(
    title='Review Tag Count per Business Type',  # Change the plot title
    xaxis_title='Business Type',  # Change the X-axis title
    yaxis_title='Review Tag Count'  # Change the Y-axis title
)

fig.show()
```

```python
#Visualization 2

# Merge the DataFrames on a common column
merged_df = pd.merge(df_r, df_b, on='place_id')

# Define the tags you want to focus on
focus_tags = ["pests", "fees", "price", "maintenance"]

# Process tags and associate them with review ratings
all_tags_with_scores = []
for index, row in merged_df.iterrows():
    if isinstance(row['reviews_tags'], str):
        tags = row['reviews_tags'].split(', ')
        for tag in tags:
            if tag in focus_tags:
                all_tags_with_scores.append({'tag': tag, 'review_rating': row['review_rating']})

# Convert to DataFrame
tags_scores_df = pd.DataFrame(all_tags_with_scores)

# Aggregate average review rating by tag
average_scores_by_tag = tags_scores_df.groupby('tag')['review_rating'].mean().reset_index()

# Visualization
fig = px.bar(average_scores_by_tag, x='tag', y='review_rating', title='Average Rating by Specific Tags')

fig.update_layout(
    title='Average Rating by Specific Tags',
    xaxis_title='Tag',
    yaxis_title='Average Rating'
)

fig.show()
```

```python
#Visualization 3

# Merge the DataFrames on a common column
merged_df = pd.merge(df_r, df_b, on='place_id')

# Process all tags
all_tags = []
for index, row in merged_df.iterrows():
    if isinstance(row['reviews_tags'], str):
        tags = row['reviews_tags'].split(', ')
        for tag in tags:
            all_tags.append(tag.strip())  # Strip whitespace

# Convert to DataFrame
tags_df = pd.DataFrame(all_tags, columns=['tag'])

# Count the occurrences of each tag and get the top 10
tag_counts = tags_df['tag'].value_counts().nlargest(10).reset_index()
tag_counts.columns = ['tag', 'count']

# Visualization using Pie Chart
fig = px.pie(tag_counts, values='count', names='tag', title='Top 10 Most Frequent Tags in Reviews')

fig.show()
```

```python
#Visualization 4

df_r['review_datetime_utc'] = pd.to_datetime(df_r['review_datetime_utc'])

# Extract year from 'review_datetime_utc'
df_r['year'] = df_r['review_datetime_utc'].dt.year

# Group by year and calculate average rating
yearly_avg = df_r.groupby('year')['review_rating'].mean().reset_index()

# Create a line chart
fig = px.line(yearly_avg, x='year', y='review_rating',
              title='Average Review Rating Over Years',
              labels={'year': 'Year', 'review_rating': 'Average Rating'})

# Show the plot
fig.show()
```

```python
merged_df = pd.merge(df_r, df_b, on='place_id')

# Count the number of reviews for each business name
review_counts = merged_df['name'].value_counts().reset_index()
review_counts.columns = ['name', 'number_of_reviews']

# Sort the businesses by review count and select the top 50
top_businesses = review_counts.nlargest(50, 'number_of_reviews')

# Create a sunburst chart for the top 50 businesses
fig = px.sunburst(top_businesses, path=['name'], values='number_of_reviews',
                  title='Top 50 Businesses with highest Number of Reviews')

# Display the plot
fig.show()
```
